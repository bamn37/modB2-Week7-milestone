Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

to bring attention to the last three months of data we can use different kinds of preattentive attributes to do that:

**color**

plot earlier months in a neutral color and highlight the last three months in a bright/contrasting color.

**line thickness**

plot earlier months in a thin line and then hightlight the last three months in a thicker line.

**markers**

use distince markers for the last three months (such as circles or stars) and no markers for first 9 months.

**annotation**

use labels or arrows to point to the final three months

**opacity**

use an alpha to make the earlier 9 months fade into background a bit

**use a shape**

use a rectangle to "highlight" the last three months by drawing a rectangle around that area.

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='ME')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='ME')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import ListedColormap
from matplotlib.patches import Patch
data = {
    'Country': ['Australia', 'Brazil', 'Canada', 'China', 'France', 'Germany', 'India', 'Italy', 'Mexico', 'Russia', 'Spain', 'Turkey', 'United Kingdom', 'United States'],
    'A': [1, 1, 2, 1, 3, 3, 4, 2, 1, 4, 2, 7, 1, 1],
    'B': [2, 3, 3, 2, 2, 1, 1, 4, 5, 3, 3, 2, 2, 2],
    'C': [3, 4, 6, 8, 4, 6, 8, 10, 4, 7, 4, 3, 3, 4],
    'D': [6, 5, 12, 4, 8, 5, 10, 9, 6, 9, 5, 4, 6, 3],
    'E': [7, 6, 8, 7, 10, 4, 5, 8, 3, 12, 11, 8, 7, 5]}

df = pd.DataFrame(data)
df.set_index('Country', inplace=True)

df_clipped = df.clip(upper=5)

colors = ["#08306b", "#2171b5", "#6baed6", "#c6dbef", "#ffffff"]
cmap = ListedColormap(colors)

plt.figure(figsize=(10,10))
ax = sns.heatmap(df_clipped, cmap=cmap, linewidths=0.5, linecolor='gray', cbar=False,
                 square=True, annot=df, fmt='d', annot_kws={"fontsize":10})

ax.set_ylabel('')

ax.xaxis.tick_top()
plt.xticks(rotation=0)

legend_labels = ["1", "2", "3", "4", "5+"]
legend_colors = ["#08306b", "#2171b5", "#6baed6", "#c6dbef", "#ffffff"]
patches = [Patch(facecolor=legend_colors[i], edgecolor='gray', label=legend_labels[i]) for i in range(len(legend_labels))]
plt.legend(handles=patches, title="Rank", bbox_to_anchor=(0.5, 1.15), loc='upper center', ncol=5)

plt.text(0.23, 0.91, "Country | Drug", fontsize=16, fontweight='bold', transform=plt.gcf().transFigure)

plt.tight_layout()
plt.show()
```
